```python
import pandas as pd
import numpy as np
# Read the CSV file into a DataFrame
df = pd.read_csv(r"C:\Users\HP\Downloads\SSA_Data.csv")
# Drop unnecessary columns
df = df.drop(["Country Name", "Country Code", "Series Code"], axis=1)
# Function to clean column names (remove text in brackets)
def clean_column_names(col):
  if "[" in col:
    return col.split("[")[0]  # Split at the first bracket and return the part before
  else:
    return col                 # Return the column name as it is if no brackets
# Apply the cleaning function to all column names
df.columns = df.columns.map(clean_column_names)
# Create a copy of the DataFrame to avoid modifying the original
df1 = df.copy()
# Select columns to convert to numeric (excluding the first column)
cols_to_convert = df1.columns[1:]
# Convert the selected columns to numeric, handling errors with 'coerce'
for col in cols_to_convert:
  df1[col] = pd.to_numeric(df1[col], errors='coerce')  # Replace non-numeric values with NaN
# Melt the DataFrame to reshape it (useful for pivoting)
df_melted = df1.melt(id_vars="Series Name",
                      var_name="Year",
                      value_name="value")
# Create a pivot table to summarize the data by Year and Series Name
df_pivoted = df_melted.pivot_table(index='Year',
                                   columns='Series Name',
                                   values='value',
                                   fill_value=0)  # Fill missing values with 0
df_pivoted.head()
```
